# ETL, Configuration

Imports

```python
import pandas as pd
import numpy as np
import sqlite3 as sql
import pyodbc as py
import warnings
warnings.simplefilter('ignore')
```

Database Connection

```python
DB = {'servername': r'MOHAMEDS\SQLEXPRESS',
      'database': 'DatawarehouseP2'}

export_conn = py.connect('DRIVER={SQL SERVER}; SERVER='+
                         DB['servername']+
                         ';DATABASE='+
                         DB['database']+
                         ';Trusted_Connection=yes'
)

export_conn = export_conn.cursor()
```

SQL (Database Connection) inladen

```python
AENC_conn = sql.connect("aenc.sqlite")
```

# Data Inlezen

AENC


```python
Department = pd.read_sql_query('Select * from Department', AENC_conn)
Product = pd.read_sql_query('Select * from Product', AENC_conn)
State = pd.read_sql_query('Select * from State', AENC_conn)
Bonus = pd.read_sql_query('Select * from Bonus', AENC_conn)
Customer = pd.read_sql_query('Select * from Customer', AENC_conn)
Employee = pd.read_sql_query('Select * from Employee', AENC_conn)
Sales_Order = pd.read_sql_query('Select * from Sales_Order', AENC_conn)
Sales_Order_Item = pd.read_sql_query('Select * from Sales_Order_Item', AENC_conn)
```

Data Joined

```python
EmployeeBonus = Bonus.merge(Employee, on='emp_id')
SalesOrder_M_OrderItem = Sales_Order_Item.merge(Sales_Order, on='id')
```

Data Insertion for AENC

```python
for index, row in Department.iterrows():
    try:
        query = f"""INSERT INTO Department VALUES(
        {row['dept_id']},
        '{row['dept_name'].replace("'", "''")}',
        {row['dept_head_id']})"""

        export_conn.execute(query)

    except Exception as e:
        print(f"Error: {e}")
        print(f"Failed Query: {query}")

export_conn.commit()
```
